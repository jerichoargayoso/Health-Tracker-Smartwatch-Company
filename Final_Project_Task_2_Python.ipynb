```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/health_spending.csv')
df.head()
```

```python
df_health_spending = df.groupby('state_usa')['avg_spending'].mean()

fig, ax = plt.subplots(figsize=(12,10))
df_health_spending.plot(kind='barh')
plt.title('Health Spending Per State')
plt.ylabel('State')
plt.xlabel('Health Spending')
plt.xticks(rotation=0)
```
